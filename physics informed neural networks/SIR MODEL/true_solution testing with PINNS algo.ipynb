```python
import sys
sys.path.insert(0, 'eslearn/')
import tensorflow as tf                                                        
import numpy as np
import matplotlib.pyplot as plt
import time
import csv
import pandas as pd

# Declaration of Global variables (X is time in Days) 
X =[]
S =[]
I =[]
R =[]

#Setup of random functions

np.random.seed(1234)
tf.set_random_seed(1234)

class PhysicsInformedNN:
    # Initialize the class
    def __init__(self, t, s, i , r , layers):
        
        X = np.concatenate([t,s,i,r], 1)
                
        self.lb = t.min()
        self.ub = t.max()
                
        self.X = X
        
        self.t = X[:,0:1]
        self.s = X[:,1:2]
        self.i = X[:,2:3]
        self.r = X[:,3:4]
        
        
        self.layers = layers
        
        # Initialize the Neural Network
        self.weights, self.biases = self.initialize_NN(layers)        
        
        # Initialize the equation parameters
        
        self.lambda_1 =  tf.Variable([0.0], dtype=tf.float32) # mu
        self.lambda_2 =  tf.Variable([0.0], dtype=tf.float32) # psi
        
        # lock the parameter within certain ranges [0 ~ 0.2]
        
#         self.lambda_1 = tf.math.tanh(self.lambda_1)*-1 + 1
#         self.lambda_2 = tf.math.tanh(self.lambda_2)*-1 + 1
    
        
        
#         self.lambda_5 =  tf.Variable([0.0], dtype=tf.float32) # Delta
        
        #Initialize the tensorflow session
        self.sess = tf.Session(config=tf.ConfigProto(allow_soft_placement=True,
                                                      log_device_placement=True))
        
        # Initialize the equation Variables
        self.t_tf = tf.placeholder(tf.float32, shape=[None, self.t.shape[1]])
        self.s_tf = tf.placeholder(tf.float32, shape=[None, self.s.shape[1]])
        self.i_tf = tf.placeholder(tf.float32, shape=[None, self.i.shape[1]])
        self.r_tf = tf.placeholder(tf.float32, shape=[None, self.r.shape[1]])
#         self.d_tf = tf.placeholder(tf.float32, shape=[None, self.d.shape[1]])
        
        
        self.s_pred, self.i_pred, self.r_pred, self.f_s_pred, self.f_i_pred,self.f_r_pred = self.net_NS(self.t_tf)
        
        #The loss function
        
        self.loss = tf.reduce_sum(tf.square(self.s_tf - self.s_pred)) + \
                    tf.reduce_sum(tf.square(self.i_tf - self.i_pred)) + \
                    tf.reduce_sum(tf.square(self.r_tf - self.r_pred)) + \
                    tf.reduce_sum(tf.square(self.f_s_pred)) + \
                    tf.reduce_sum(tf.square(self.f_i_pred)) + \
                    tf.reduce_sum(tf.square(self.f_r_pred))


#         self.loss = tf.reduce_mean(tf.square(self.s_tf - self.s_pred)) + \
#                     tf.reduce_mean(tf.square(self.i_tf - self.i_pred)) + \
#                     tf.reduce_mean(tf.square(self.r_tf - self.r_pred)) + \
#                     tf.reduce_mean(tf.square(self.f_s_pred)) + \
#                     tf.reduce_mean(tf.square(self.f_i_pred)) + \
#                     tf.reduce_mean(tf.square(self.f_r_pred))
        
        # The optimizer function            
        self.optimizer = tf.contrib.opt.ScipyOptimizerInterface(self.loss, 
                                                                method = 'L-BFGS-B', 
                                                                options = {'ftol' : 1.0 * np.finfo(float).eps})        
        
        self.optimizer_Adam = tf.train.AdamOptimizer()
        self.train_op_Adam = self.optimizer_Adam.minimize(self.loss)                    
        
        init = tf.global_variables_initializer()
        self.sess.run(init)
    
    #The neural network
    def initialize_NN(self, layers):        
        weights = []
        biases = []
        
        num_layers = len(layers) 
        for l in range(0,num_layers-1):
            W = self.xavier_init(size=[layers[l], layers[l+1]])
            b = tf.Variable(tf.zeros([1,layers[l+1]], dtype=tf.float32), dtype=tf.float32)
            weights.append(W)
            biases.append(b)
           
        return weights, biases
        
    def xavier_init(self, size):
        in_dim = size[0]
        out_dim = size[1]        
        xavier_stddev = np.sqrt(2.0/(in_dim + out_dim))
        return tf.Variable(tf.truncated_normal([in_dim, out_dim], stddev = xavier_stddev), dtype=tf.float32)
    
    def neural_net(self, X, weights, biases):
        num_layers = len(weights) + 1        
        H = 2.0*(X - self.lb)/(self.ub - self.lb) - 1.0
        for l in range(0,num_layers - 2):
            W = weights[l]
            b = biases[l]
            H = tf.tanh(tf.add(tf.matmul(H, W), b))
        W = weights[-1]
        b = biases[-1]
        Y = tf.add(tf.matmul(H, W), b)
        return Y
        
    def net_NS(self, t):
        lambda_1 = self.lambda_1       
        lambda_2 = self.lambda_2
        
        psi_and_p = self.neural_net(tf.concat([t], 1), self.weights, self.biases)
        
        S = psi_and_p[:,0:1]
        I = psi_and_p[:,1:2]
        R = psi_and_p[:,2:3]
        
        
        s_t = tf.gradients(S, t)[0]
        i_t = tf.gradients(I, t)[0]
        r_t = tf.gradients(R, t)[0]
    

#         #The Residual functions
       
#         f_s = s_t +  lambda_1*(S*I) 
#         f_i = i_t -  lambda_1*(S*I) + lambda_3*I + lambda_5*I
#         f_r = r_t  - lambda_3*I 
#         f_d = d_t - lambda_5*I
        
#         return S, I, R, D, f_s, f_i, f_r, f_d
        
        N = float(567169)
        beta = float(0.000006)
        delta = float(0.00002)
        rho = float(0)
        ab = float(0)
        phi = float(3.0173)
        
        # mu = lambda_1
        # psi = lambda_2
        
        
        f_s = s_t - beta*N + lambda_1*S*I/N - ab*R + delta*S
        f_i = i_t -  lambda_1*(S*I)/N + (lambda_2 + lambda_1)*I + ((rho*I)/(phi + I))
        f_r = r_t  - lambda_2*I - ((rho*I) / (phi + I)) + (ab + delta)*R
        
        return S, I, R, f_s, f_i, f_r
    
    def callback(self,loss,lambda_1, lambda_2):
        print('Loss: %.3e, l1: %.7f, l2: %.7f' % (loss,lambda_1, lambda_2))
      
    def train(self, nIter): 

        tf_dict = {self.t_tf: self.t,self.s_tf: self.s, self.i_tf: self.i,self.r_tf: self.r}
        
        start_time = time.time()
        for it in range(nIter):
            self.sess.run(self.train_op_Adam, tf_dict)
            
            # Print
            if it % 10 == 0:
                elapsed = time.time() - start_time
                loss_value = self.sess.run(self.loss, tf_dict)
                lambda_1_value = self.sess.run(self.lambda_1)
                lambda_2_value = self.sess.run(self.lambda_2)
                
                
                print('It: %d, Loss: %.3e, l1: %.7f, l2: %.7f, Time: %.2f' % 
                      (it, loss_value, lambda_1_value,lambda_2_value, elapsed))
                start_time = time.time()
            
        self.optimizer.minimize(self.sess,
                                feed_dict = tf_dict,
                                fetches = [self.loss,self.lambda_1,self.lambda_2],
                                loss_callback = self.callback)
            
    
    def predict(self, t_star):
        
        tf_dict = {self.t_tf: t_star}
        
        s_star = self.sess.run(self.s_pred, tf_dict)
        i_star = self.sess.run(self.i_pred, tf_dict)
        r_star = self.sess.run(self.r_pred, tf_dict)
        return s_star, i_star, r_star

#The main fuction of the program

if __name__ == "__main__": 
     
    N = 100
    
    layers = [1,128,128,128, 3]
    #layers = [1,30,30,30,3]

    
#     with open('data.csv', mode='r') as csv_file:
#         csv_reader = csv.DictReader(csv_file)
#         line_count = 0
 
#         #Read data from the CSV file
#         for row in csv_reader:
#             X.append(row['day'])
#             S.append(row['susceptible'])
#             I.append(row['infected'])
#             R.append(row['recovered'])

    df = pd.read_csv('true_solution.csv')
    X.append(df['Day'])
    S.append(df['Susceptible'])
    I.append(df['Infected'])
    R.append(df['Recovered'])

    # Find total size of data
    total = np.size(X)
    
    #Reshaping the arrays of Data
    X = np.reshape(X,(total,1))
    S = np.reshape(S,(total,1))
    I = np.reshape(I,(total,1)) 
    R = np.reshape(R,(total,1)) 
    
    #Make the array of data float type
    XX = np.array(X,dtype=float)
    SS = np.array(S,dtype=float)
    II = np.array(I,dtype=float)
    RR = np.array(R,dtype=float)
    
    
    #Non-dimentionalize the data
#     XX = np.true_divide(XX, 1)
#     SS = np.true_divide(SS, 762)
#     II = np.true_divide(II, 762)
#     RR = np.true_divide(RR, 762)
         
# Training Data    
    idx = np.random.choice(N,N, replace=False)
    
#     t_train = XX[336:367]
#     S_train = SS[336:367]
#     I_train = II[336:367]
#     R_train = RR[336:367]  
    
    t_train = XX[300:400]
    S_train = SS[300:400]
    I_train = II[300:400]
    R_train = RR[300:400] 
    
    #model = PhysicsInformedNN(t_train,S_train,I_train,R_train,layers) \
    model = PhysicsInformedNN(XX,SS,II,RR,layers) 
    start_time = time.time()
    model.train(1000)
    stop = time.time() - start_time
    
    print('Elapsed %.3f seconds' %stop)
    

```
